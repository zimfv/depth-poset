```python
import numpy as np
import pandas as pd

import itertools
import inspect

from src import poset_scores
from src import node_scores
```

# Extended Cubical Torus

```python
poset_score_names = [name for name, obj in inspect.getmembers(poset_scores, inspect.isfunction) if obj.__module__ == poset_scores.__name__]

df_poset_scores = pd.DataFrame(False, index=poset_score_names, columns=pd.MultiIndex.from_tuples(itertools.product([1, 2, 3], ['full', 'column reduction', 'row reduction']), names=['dim', 'object']))

df_poset_scores.loc[['number_of_nodes', 'number_of_relations', 'cycles_dimension', 'height'], df_poset_scores.columns.get_level_values(1) == 'full'] = True
df_poset_scores.loc[['number_of_components', 'number_of_maximal_nodes', 'number_of_minimal_nodes'], 2] = True

df_poset_scores.to_csv('params/extended_cubical_torus_interesting_poset_scores.csv')

df_poset_scores
```

```python
node_score_names = [name for name, obj in inspect.getmembers(node_scores, inspect.isfunction) if obj.__module__ == node_scores.__name__]

df_node_scores = pd.DataFrame(False, index=node_score_names, columns=pd.MultiIndex.from_tuples(itertools.product([1, 2, 3], ['full', 'column reduction', 'row reduction']), names=['dim', 'object']))

#df_node_scores.loc[['ancestors_number', 'successors_number']] = True

df_node_scores.to_csv('params/extended_cubical_torus_interesting_node_scores.csv')

df_node_scores
```
